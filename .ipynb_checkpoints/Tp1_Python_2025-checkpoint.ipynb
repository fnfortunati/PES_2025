# Ejercicio 1 - Filtro Moving Average

1. Implementá un algoritmo de promedio móvil (moving average) según la definición vista en clase, que reciba como parámetros:

- La señal de entrada.
- El tamaño de la ventana.

A partir de tu implementación, obtené y graficá la respuesta impulsiva del sistema (lo que en clase llamamos su “firma” o función de transferencia).

2. Respondé:

- ¿Cómo se obtiene la respuesta impulsiva de cualquier sistema?
  
  **Aplicandole un impulso al sistema del que pretendo obtener la respuesta.**
  
- ¿Qué representa en el caso del filtro de promedio móvil?

  **Partiendo de que "M+N+1" es la ventana (V). La respuesta impulsiva de un filtro de promedio móvil es una secuencia rectangular   de amplitud 1/V y longitud V. Esto refleja que el filtro toma un promedio de V muestras pasadas en cada instante.**

**Partiendo de la siguiente fórmula:**

$$y[n] = {1 \over {M+N+1}} \sum_{k=-M}^N x[n-k]$$

```python
import numpy as np
import matplotlib.pyplot as plt

def prom_mov(sig, vent):
    M = vent // 2
    N = M + 1
    sal = []
    
    for i in range(len(sig)):
        suma = 0
        val = 0
        for k in range(-M, N):
            idr = i + k
            if 0 <= idr < len(sig):
                suma += sig[idr]
                val += 1
        sal.append(suma / val)
    return sal

# Parámetros
vent = 5   # tamaño de ventana
L = 20     # longitud para visualizar

# Delta discreto
delta = np.zeros(L)
delta[L//2] = 1   # delta centrado en el medio

# Respuesta al impulso
h = prom_mov(delta, vent)

# Graficar
plt.stem(range(L), h)
plt.title(f"Respuesta al impulso del promedio móvil (ventana={vent})")
plt.xlabel("n")
plt.ylabel("h[n]")
plt.grid(True)
plt.show()
```

# Ejercicio 2 - Respuesta del sistema

1. Genera dos señales de prueba:
    - Una onda cuadrada de amplitud 1 y frecuencia de 2 kHz.
    - Una onda triangular de amplitud 1 y frecuencia de 1 kHz.

Aplica el moving average implementado en el Ejercicio 1 a estas señales.

2. Analiza los resultados:

    - Explica qué cambios observas en la forma de onda.

3. Escribe tus conclusiones: ¿qué efecto tiene el filtro de promedio móvil sobre cada señal?

    **Se aprecia que a medida que la ventana es mayor, la señal filtrada pierde relacion con la señal original.**

    **Suponiendo una ventana de 10 muestras, de forma de perder la amplituda de la señal original:**

    **En la cuadrada se puede observar que la señal filtrada se asemeja mas a una triangular; mientras que en la señal triangular, la señal filtrada se asemeja a una senoidal.**


```python
import scipy.signal as sig

# Parámetros
frec_muestreo = 50000   #Muestreo mucho mas alto al criterio de Nyquist, para obtener una buena señal cuadrada
frec_cuad = 2000        
frec_trg = 1000         

# Vector de tiempo
t = np.arange(0, 1, 1/frec_muestreo)

#Ventana del promedio movil
vent = 10

# Señales
cuad = sig.square(2 * np.pi * frec_cuad * t)
trg = sig.sawtooth(2 * np.pi * frec_trg * t, width=0.5)

#Señales Filtradas
cuad_avg = prom_mov(cuad, vent)
trg_avg = prom_mov(trg, vent)

# --- Gráfico señal Cuadrada  ---
T_cuad = 1 / frec_cuad      # período de la onda cuadrada
plt.figure(figsize=(10,4))
plt.title('Señal Cuadrada')
plt.plot(t, cuad)
plt.xlim(0, 3*T_cuad)
plt.ylim(-2, 2)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# --- Gráfico señal Cuadrada Filtrada ---
plt.figure(figsize=(10,4))
plt.title('Señal Cuadrada Filtrada')
plt.plot(t, cuad_avg)
plt.xlim(0, 3*T_cuad)
plt.ylim(-2, 2)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# --- Gráfico señal Triangular ---
T_trg = 1 / frec_trg        # período de la onda triangular
plt.figure(figsize=(10,4))
plt.title('Señal Triangular')
plt.plot(t, trg)
plt.xlim(0, 3*T_trg)
plt.ylim(-2, 2)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# --- Gráfico señal Triangular Filtrada ---
plt.figure(figsize=(10,4))
plt.title('Señal Triangular Filtrada')
plt.plot(t, trg_avg)
plt.xlim(0, 3*T_trg)
plt.ylim(-2, 2)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

# Ejercicio 3 - Convolución en punto flotante y punto fijo

1. Implementa un algoritmo de convolución:

    - En punto flotante (float).
    - En punto fijo, utilizando por ejemplo la librería fixedpoint.

2. Compara los resultados con la función de NumPy np.convolve. Mide y compara:

    - El tiempo de ejecución usando timeit.

3. Conclusión:

    - ¿Qué diferencias observas entre las implementaciones?
       **De acuerdo a la salida, no se aprecia una diferencia sustancial en las implementaciones. En cuanto a tiempo, utilizar la la función de Numpy trae resultados mucho más rápidos.**
      
    - ¿Qué ventajas y desventajas tiene usar punto fijo en lugar de punto flotante?
      **Según la explicación en clase y de acuerdo a material de investigación; utilizar punto fijo debería ser mucho mas provechoso que usar punto flotante. Según es análisis que se hizo de los tiempos de las implementaciones, esto no se evidencia.**
  

      **De acuerdo al material consultado, las ventajas y desventajas de usar punto fijo en vez de punto flotante son:**
  
      1. Ventajas:
     
         - Mayor eficiencia en recursos computacionales.
         - Menor uso de memoria.
         - Tiene menos costo de Hardware.

      2. Desventajas:

         - Menor rango dinámino.
         - Menor precisión.
         
     
           

```python
from fixedpoint import FixedPoint

def convolucion_float(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0.0] * len_y
    for n in range(len_y):
        acc = 0.0
        for k in range(len_h):
            if 0 <= n - k < len_x:
                acc += x[n - k] * h[k]
        y[n] = acc
    return y

def convolucion_punto_fijo(x, h, qformat='s16.16'):
    x_fp = [FixedPoint(val, qformat) for val in x]
    h_fp = [FixedPoint(val, qformat) for val in h]
    
    len_x, len_h = len(x_fp), len(h_fp)
    len_y = len_x + len_h - 1
    y_fp = [FixedPoint(0, qformat) for _ in range(len_y)]
    
    for k, h_val in enumerate(h_fp):
        for n, x_val in enumerate(x_fp):
            y_fp[n + k] += x_val * h_val  # sumamos directamente a la posición correcta
    return y_fp

# Parámetros de los impulsos cuadrados
t = np.linspace(0, 1, 1000, endpoint=False)  # Vector de tiempo
impulso1 = np.zeros_like(t)
impulso1[200:300] = 1  # Impulso cuadrado 1
impulso2 = np.zeros_like(t)
impulso2[400:500] = 1  # Impulso cuadrado 2

# Realizar la convolución de los impulsos
sig_conv_flt = convolucion_float(impulso1,impulso2)
sig_conv_fijo = convolucion_punto_fijo(impulso1,impulso2)
sig_conv_np = np.convolve(impulso1, impulso2)


# Convertir señal de punto fijo a float para graficar
sig_conv_fijo = [float(val) for val in sig_conv_fijo]

# --- Gráfico Convolucion Punto Flotante ---
plt.figure(figsize=(10,4))
plt.title('Convolucion Float vs Fijo')
plt.plot(np.arange(len(sig_conv_flt))/1000, sig_conv_flt, label='Punto flotante', color='blue', linewidth=2)
plt.plot(np.arange(len(sig_conv_fijo))/1000, sig_conv_fijo, label='Punto fijo', color='red', linestyle='--', linewidth=2)
plt.plot(np.arange(len(sig_conv_np))/1000, sig_conv_np, label='Punto fijo', color='green', linestyle=':', linewidth=2)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

```python
# Ejemplos de medicion de tiempos
# %pylab inline
import timeit
import time
import numpy as np

t = np.arange(0, 1024/20000, 1/20000)
s1= np.sin(2*np.pi*1000*t)+1
s2 = np.sin(2*np.pi*1200*t)

# METODO 1 de medición de tiempos

# Este wrapper empaqueta funciones para que el timeit pueda medir solo eso
def wrapper(func, *args):
    def wrapped():
        return func(*args)
    return wrapped

conv = wrapper(np.convolve, s1, s2)
print (timeit.timeit(conv, number=1))

# METODO 2 de medición de tiempos
start = time.time()
conv_r = np.convolve(s1,s2)
end = time.time()
print (end-start)
```

# Análisis de Tiempos de la Convolución

```python
time_float = timeit.timeit(wrapper(convolucion_float, impulso1, impulso2), number=1)
time_fixed = timeit.timeit(wrapper(convolucion_punto_fijo, impulso1, impulso2), number=1)
time_numpy = timeit.timeit(wrapper(np.convolve, impulso1, impulso2), number=1)

print("Tiempo convolución float:", time_float)
print("Tiempo convolución punto fijo:", time_fixed)
print("Tiempo np.convolve:", time_numpy)
```
